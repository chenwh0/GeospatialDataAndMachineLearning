## Key Concepts to Remember

**The Four GRIDMET Drought Indices:**

**1. Standardized Precipitation Index (SPI)**
- Equation: $SPI = \Phi^{-1}(G(P))$ where $\Phi^{-1}$ is inverse normal distribution
- Measures: Precipitation anomalies from long-term mean
- Strength: Simple, directly interpretable, good for agricultural drought
- Limitation: Doesn't account for temperature effects
- Typical range: -3 to +3 (standard deviations)

**2. Standardized Precipitation Evapotranspiration Index (SPEI)**
- Equation: Water balance $D_i = P_i - PET_i$ then standardized
- Measures: Precipitation minus potential evapotranspiration
- Strength: Includes temperature effects, sensitive to climate change
- Limitation: Requires good PET estimates
- Typical range: -3 to +3

**3. Evaporative Demand Drought Index (EDDI)**
- Equation: $EDDI = \frac{PET_i - \overline{PET}}{\sigma_{PET}}$
- Measures: Atmospheric evaporative demand anomalies
- Strength: Early warning signal before precipitation deficits manifest
- Limitation: Complements but doesn't replace precipitation indices
- Typical range: -2 to +2

**4. Palmer Drought Severity Index (PDSI)**
- Complex calculation involving two-layer soil water balance model
- Measures: Persistent soil moisture departures from normal
- Strength: Long-term memory, captures drought persistence
- Limitation: Complex calculation, different scale than other indices
- Typical range: -6 to +6

**Timescale Selection:**
- 14-day: Real-time stress, week-to-week variability
- 30-day: Monthly patterns, operational decisions
- 90-day: Seasonal drought characteristics, crop impacts
- 180-day/1-year: Long-term patterns, water resource planning
- Multi-year timescales: Persistent vs. transient conditions

**Drought Type Classification:**
- **Meteorological**: Precipitation deficits (measured by SPI)
- **Agricultural**: Soil moisture stress during growing season (measured by SPEI-30d to 90d)
- **Hydrological**: Reduced water supplies (measured by PDSI, SPEI-180d to 1y)
- **Socioeconomic**: Water shortage impacts (requires multiple indices + context)
