## Task 2: Comprehensive Multi-Index Fire Analysis

### Instructions:

Following the advanced workflows demonstrated in class materials, create **five comprehensive analyses** integrating multiple fire detection and vegetation indices. You must adapt the code examples to your specific fire event.

### Required Analyses:

**Analysis A: Pre-Fire and Post-Fire dNBR Assessment**

- Use Sentinel-2 (preferred) or Landsat 8/9 imagery
- Calculate NBR (Normalized Burn Ratio) for pre-fire and post-fire periods
  - Sentinel-2: NBR = (B8 - B12) / (B8 + B12)
  - Landsat: NBR = (SR_B5 - SR_B7) / (SR_B5 + SR_B7)
- Generate dNBR = Pre-Fire NBR - Post-Fire NBR
- Apply USGS burn severity classification (multiply dNBR by 1000):
  - Unburned/Very Low: < 100
  - Low Severity: 100-270
  - Moderate-Low Severity: 270-440
  - Moderate-High Severity: 440-660
  - High Severity: ≥ 660
- **CRITICAL:** Apply proper cloud masking to both pre-fire and post-fire composites
- Create spatial burn severity map showing all five severity classes

**Analysis B: Advanced Burn Index Assessment (BAI, BAIS2, MIRBI)**

- Calculate post-fire burn-specific indices on your post-fire composite:
  - **BAI** (Burned Area Index): $BAI = \frac{1}{(0.1 - RED)^2 + (0.06 - NIR)^2}$
  - **BAIS2** (for Sentinel-2): $BAIS2 = (1 - \sqrt{\frac{B06 \times B07 \times B8A}{B04}}) \times (\frac{B12 - B8A}{\sqrt{B12 + B8A}} + 1)$
  - **MIRBI**: $MIRBI = 10 \times SWIR2 - 9.8 \times SWIR1 + 2$
- Identify burned areas using index-specific thresholds:
  - BAI > 100 indicates burn signature
  - BAIS2 > 0.05 indicates fire effects
  - MIRBI > 3.5 indicates thermal alteration
- Compare burn detection patterns across all three indices
- Create maps showing burned areas detected by each index

**Analysis C: Vegetation Health Multi-Index Analysis**

- Calculate vegetation indices for both pre-fire and post-fire periods:
  - **NDVI** = (NIR - Red) / (NIR + Red)
  - **NDMI** = (NIR - SWIR1) / (NIR + SWIR1)
  - **EVI** = $2.5 \times \frac{NIR - RED}{NIR + 6 \times RED - 7.5 \times BLUE + 1}$
- Generate change detection maps (dNDVI, dNDMI, dEVI)
- Analyze how each index captures different aspects of fire impact:
  - NDVI: vegetation greenness loss
  - NDMI: moisture content change
  - EVI: enhanced vegetation sensitivity
- Create composite visualization showing all three vegetation change patterns

**Analysis D: Multi-Index Composite Fire Severity**

- Integrate all indices into a weighted composite severity metric:
  - Composite = 0.40 × dNBR + 0.20 × dNDVI + 0.15 × dNDMI + 0.10 × dEVI + 0.10 × (postBAI - 100)/1000 + 0.05 × postBAIS2
- Normalize composite severity to 0-1 scale
- Classify into management-relevant severity categories:
  - Unburned: < 0.1
  - Low: 0.1-0.3
  - Moderate-Low: 0.3-0.5
  - Moderate-High: 0.5-0.7
  - High: ≥ 0.7
- **Create fire detection confidence map:** Areas where multiple indices agree on burn detection
  - High confidence: dNBR > 0.1 AND dNDVI > 0.05 AND postBAI > 50
- **Generate recovery potential map:** Pre-fire health + inverse severity
  - Recovery Potential = 0.4 × pre-NDVI + 0.3 × pre-NDMI + 0.3 × (1 - normalized severity)

## Key Concepts to Remember

### The Eight Fire Monitoring Indices:

**1. Normalized Burn Ratio (NBR)**
- Formula: $NBR = \frac{NIR - SWIR2}{NIR + SWIR2}$
- Measures: Vegetation health and moisture based on NIR/SWIR contrast
- Strength: Standard burn detection, widely validated
- Application: Pre/post-fire comparison (dNBR) for severity classification
- Typical range: -1 to +1

**2. Differenced NBR (dNBR)**  
- Formula: $dNBR = NBR_{pre} - NBR_{post}$
- Measures: Magnitude of vegetation and moisture change from fire
- Strength: Quantifies burn severity, USGS-validated thresholds
- Application: Primary severity classification metric
- Typical severe fire values: 0.4 to 0.7

**3. Burned Area Index (BAI)**
- Formula: $BAI = \frac{1}{(0.1 - RED)^2 + (0.06 - NIR)^2}$
- Measures: Spectral distance to charcoal signature
- Strength: Emphasizes burn area detection over severity gradients
- Application: Binary burned/unburned classification
- Typical burn threshold: BAI > 100

**4. Burned Area Index for Sentinel-2 (BAIS2)**
- Formula: $BAIS2 = (1 - \sqrt{\frac{B06 \times B07 \times B8A}{B04}}) \times (\frac{B12 - B8A}{\sqrt{B12 + B8A}} + 1)$
- Measures: Multi-band burn signature optimized for Sentinel-2
- Strength: Uses red-edge bands for enhanced burn discrimination
- Application: High-resolution burn mapping with Sentinel-2
- Typical burn threshold: BAIS2 > 0.05

**5. Mid-Infrared Burn Index (MIRBI)**
- Formula: $MIRBI = 10 \times SWIR2 - 9.8 \times SWIR1 + 2$
- Measures: Thermal alteration in SWIR bands
- Strength: Effective for low-severity burns, complements NBR
- Application: Detection of subtle burn signals
- Typical burn threshold: MIRBI > 3.5

**6. Normalized Difference Vegetation Index (NDVI)**
- Formula: $NDVI = \frac{NIR - RED}{NIR + RED}$
- Measures: Photosynthetic vegetation greenness
- Strength: Direct measure of vegetation loss from fire
- Application: Pre/post-fire comparison (dNDVI) for vegetation impact
- Typical range: -1 to +1

**7. Normalized Difference Moisture Index (NDMI)**
- Formula: $NDMI = \frac{NIR - SWIR1}{NIR + SWIR1}$
- Measures: Vegetation water content
- Strength: Sensitive to canopy moisture stress and loss
- Application: Pre/post-fire comparison (dNDMI) for moisture impact
- Typical range: -1 to +1

**8. Enhanced Vegetation Index (EVI)**
- Formula: $EVI = 2.5 \times \frac{NIR - RED}{NIR + 6 \times RED - 7.5 \times BLUE + 1}$
- Measures: Vegetation health with atmospheric correction
- Strength: Reduced saturation in dense canopies, atmospheric compensation
- Application: Pre/post-fire comparison (dEVI) for vegetation change
- Typical range: -1 to +1

### Multi-Index Integration Principles:

**Why Use Multiple Indices?**
1. **Reduces False Positives:** Shadows, water, clouds can affect single indices
2. **Captures Multiple Fire Dimensions:** Burn signature, vegetation loss, moisture change, thermal effects
3. **Increases Confidence:** Multi-index agreement validates detections
4. **Improves Severity Discrimination:** Different indices sensitive to different severity levels
5. **Supports Validation:** Cross-index comparison reveals data quality issues

**Composite Severity Weighting:**
- 40% dNBR: Primary change detection signal
- 20% dNDVI: Vegetation loss component
- 15% dNDMI: Moisture change component
- 10% dEVI: Enhanced vegetation component
- 10% BAI: Burn area signature
- 5% BAIS2: Sentinel-2 specific signal

**Fire Detection Confidence:**
- High: Multiple indices agree (e.g., dNBR > 0.1 AND dNDVI > 0.05 AND BAI > 50)
- Moderate: Some indices indicate fire, others ambiguous
- Low: Single index detection without corroboration

**Recovery Potential Assessment:**
- High: Strong pre-fire health + low-moderate severity
- Moderate: Moderate pre-fire health or moderate-high severity
- Low: Poor pre-fire health + high severity (requires intervention)

### Satellite Platform Selection:

**Sentinel-2 Advantages:**
- 10-20m resolution (higher spatial detail)
- Red-edge bands enable BAIS2 calculation
- 5-day revisit with twin satellites
- Better for small fires and perimeter detail

**Landsat 8/9 Advantages:**
- Longer temporal archive (1982+)
- 30m resolution adequate for large fires
- More consistent band calibration
- Better for temporal studies and historical context

### Burn Severity Classification Context:

**USGS Thresholds (dNBR × 1000):**
- Enhanced Regrowth: < -100 (vegetation increased post-fire)
- Unburned: -100 to 100
- Low Severity: 100 to 270 (surface fire, little tree mortality)
- Moderate-Low: 270 to 440 (mixed effects, partial canopy loss)
- Moderate-High: 440 to 660 (significant canopy mortality)
- High Severity: ≥ 660 (complete stand replacement)
